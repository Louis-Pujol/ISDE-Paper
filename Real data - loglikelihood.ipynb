```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import fcsparser
from sklearn.preprocessing import StandardScaler
import time
import pandas as pd
import pickle
from sklearn.mixture import GaussianMixture
from scipy.stats import multivariate_normal

import isde
import isde.estimators

from includes.fde import *
from includes.gaussian_mixture_cv import *

exp_name = "Levine32_N20000"

X_levine13 = np.load("data/Levine13/X.npy")
X_levine32 = np.load("data/Levine32/X.npy")

def logdensity_from_partition(X_grid, X_eval, partition, by_subsets):
    logdensity_isde = np.zeros(X_eval.shape[0])
    for S in partition:
        loc_d = len(S)
        loc_h = by_subsets[tuple(S)]['params']['bandwidth']
        logdensity_isde += isde.estimators.GaussianKDE(bandwidth=loc_h).score_samples(grid_points=X_grid[:, S], eval_points=X_eval[:, S])

    return np.ma.masked_invalid(logdensity_isde)

N_train = 5000

np.random.seed(0)
N, d = X_levine13.shape
indices = np.array(list(range(N)))
np.random.shuffle(indices)
X_levine13_train = X_levine13[indices[0:N_train], :]
X_levine13_validation = X_levine13[indices[N_train::], :]


N, d = X_levine32.shape
indices = np.array(list(range(N)))
np.random.shuffle(indices)
X_levine32_train = X_levine32[indices[0:N_train], :]
X_levine32_validation = X_levine32[indices[N_train::], :]
```

##Â Levine13

```python
print("CVKDE...")
hs = np.logspace(-3,-1,40)
%time bandwidth_kde = isde.estimators.CVKDE(X_levine13_train, params={"hs": hs, "n_fold" : 5})[0].bandwidth

print("ISDE_CVKDE...")
m, n = 3000, 2000
hs = np.logspace(-3,-1,30)
estimator = isde.estimators.CVKDE
k = 5
%time by_subsets = isde.ISDE(X=X_levine13_train, m=m, n=n, k=k, multidimensional_estimator=estimator, hs=hs, do_optimization=False, verbose=True)

print("FDE...")
graph_FDE = FDE(X_levine13_train, bw_bysubsets=True, by_subsets=by_subsets)

print("GM...")
gm = GaussianMixture_cv(X_levine13_train, min_components=1, max_components=30, criterion='cross_val')

means = gm.means_
covariances = gm.covariances_
proportions = gm.weights_

#Save
pickle.dump(by_subsets , open("data/Levine13/by_subets", "wb"))
np.save("data/Levine13/graoh_FDE.npy", graph_FDE)
np.save("data/Levine13/proportions.npy", proportions)
np.save("data/Levine13/means.npy", means)
np.save("data/Levine13/covariances.npy", covariances)
np.save("data/Levine13/bandwidth_kde.npy", bandwidth_kde)
```

```python
by_subsets = pickle.load(open("data/Levine13/by_subets", "rb"))
graph_FDE = np.load("data/Levine13/graoh_FDE.npy")
proportions = np.load("data/Levine13/proportions.npy")
means = np.load("data/Levine13/means.npy")
covariances = np.load("data/Levine13/covariances.npy")
bandwidth_kde = float(np.load("data/Levine13/bandwidth_kde.npy"))

partition = isde.find_optimal_partition(scores_by_subsets=by_subsets, max_size=5,
                                        min_size=1, exclude = [], sense='maximize')[0]

M = 20
N_v = 2000

N_valid, d = X_levine13_validation.shape
indices = list(range(N_valid))

ll_isdes = []
ll_fdes = []
ll_mixmods = []
ll_kdes = []

for i in range(M):
    np.random.shuffle(indices)
    X_v = X_levine13_validation[indices[0:N_v], :]
    
    #KDE
    logdensity_kde = isde.estimators.GaussianKDE(bandwidth=bandwidth_kde).score_samples(grid_points=X_levine13_train, eval_points=X_v)
    ll_kdes.append(logdensity_kde.mean())
    
    # ISDE
    logdensity_isde = logdensity_from_partition(X_grid=X_levine13_train, X_eval=X_v,
                                                partition=partition, by_subsets=by_subsets)
    
    ll_isdes.append(logdensity_isde.mean())
    
    # GM
    selected_nc = len(proportions)
    density_mixmod = np.zeros(N_v)
    for i in range(selected_nc):
        density_mixmod += proportions[i] * multivariate_normal(mean=means[i], cov=covariances[i]).pdf(X_v)
    ll_mixmods.append(np.mean(np.log(density_mixmod)))
    
    # FDE
    tmp = np.ma.masked_invalid(logdensity_from_tree(X_levine13_train, X_v, graph_FDE, bw_bysubsets=True, by_subsets=by_subsets))
    ll_fdes.append(tmp.mean())
    

mpl.rcParams['figure.dpi'] = 300

df = pd.DataFrame()
df["ISDE_CVKDE"] = ll_isdes
df["FDE"] = ll_fdes
df["GM"] = ll_mixmods
df["CVKDE"] = ll_kdes
sns.set_style("whitegrid")
ax = sns.boxplot(data=df)
plt.ylabel("log-likelihood")
plt.show()

sns.set_style("whitegrid")
ax = sns.boxplot(data=df)
plt.ylabel("log-likelihood")
plt.savefig("data/Levine13/results.png")
plt.clf()
```
